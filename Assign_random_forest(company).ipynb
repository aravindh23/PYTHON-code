```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv("C:/Users/ARAVINDH/Downloads/Company_Data.csv")
df.head(3)
```

```python
df.loc[df.Sales >=8.5,'sales']= 'high'
df.loc[df.Sales <8.5,'sales']= 'low'
```

```python
df.drop('Sales',1,inplace=True)
```

```python
df.head(3)
```

```python
df.info()
```

```python
label_encoder = preprocessing.LabelEncoder() 
df['US']= label_encoder.fit_transform(df['US']) 
df['Urban']= label_encoder.fit_transform(df['Urban']) 
df['ShelveLoc']= label_encoder.fit_transform(df['ShelveLoc'])
```

```python
colnames=list(df.columns)
colnames
```

```python
predictors=colnames[0:10]
target=colnames[10]
```

```python
predictors
```

```python
target
```

```python
df.head(2)
```

```python
label_encoder = preprocessing.LabelEncoder() 
df['sales']= label_encoder.fit_transform(df['sales']) 
```

```python
x=df[predictors]
y=df[target]
```

```python
random=RandomForestClassifier(n_jobs=2,oob_score=True,n_estimators=1000,criterion="entropy")
random.fit(x,y)
```

```python
random.estimators_
```

```python
df['prediction']=random.predict(x)
```

```python
df.head(2)
```

```python
#Confusion Matrix
```

```python
confusion_matrix(df['sales'],df['prediction'])
```

```python
#BAGGING
```

```python
seed = 3
```

```python
kfold = KFold(n_splits=10, random_state=seed)
```

```python
cart = DecisionTreeClassifier()
```

```python
num_trees = 10
```

```python
model=BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=seed)
```

```python
results=cross_val_score(model,x,y,cv=kfold)
```

```python
print(results.mean())
```
